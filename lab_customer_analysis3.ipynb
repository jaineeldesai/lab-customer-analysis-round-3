```python
group_sales=data.groupby(['Sales Channel'])
#group['Response'].value_counts().plot(kind='bar')
```

```python
group_sales['Response'].value_counts().unstack().plot(kind='bar')
plt.ylabel('Response Count')
plt.title('Response Count vs Sales Channel')
plt.show()
```

```python
group_per=group_sales['Response'].value_counts('Yes').unstack()*100
group_per.plot(kind='bar')
plt.ylabel('Response %')
plt.title('Response Rate vs Sales Channel')
plt.show()
```
